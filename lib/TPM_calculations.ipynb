```python
import counts
from settings import *
from collections import defaultdict
config_dict = process_config()
gff = config_dict["ref"]["gff"]
patric_gff = config_dict["ref"]["patric_gff"]
info_file = config_dict["data"]["exp_design"]
info = pd.read_csv(info_file, index_col=0)
```

```python
pids = defaultdict(lambda: x, counts.parse_patric_gff(patric_gff))
```

```python
# 4 hour time point
four_hour_ids = info[info['experiment'] == "E1"].index
counts_dir = config_dict["counts"]["counts_dir_4"]
tpms = counts.normalize_counts_to_tpm(counts_dir, patric_gff)
df = pd.DataFrame.from_dict(tpms)
df.columns = ["S"+c for c in df.columns]
df = df[four_hour_ids]
df["tags"] = [pids[f] for f in df.index]
df = df.set_index("tags")
fn = os.path.join(counts_dir, "TPMs.csv")
df.to_csv(fn)
df.head()
```

```python
# 30 min time point 
thirty_min_ids = info[info['experiment'] != "E1"].index
counts_dir = config_dict["counts"]["counts_dir_30"]
tpms = counts.normalize_counts_to_tpm(counts_dir, patric_gff)
df = pd.DataFrame.from_dict(tpms)
df.columns = ["S"+c for c in df.columns]
df = df[thirty_min_ids]
df["tags"] = [pids[f] for f in df.index]
df = df.set_index("tags")
fn = os.path.join(counts_dir, "TPMs.csv")
df.to_csv(fn)
df.head()
```
