```python
import counts
import pandas as pd
import os
```

```python
pids = counts.parse_patric_gff(gff)
```

```python
# 4 hour time point
cf = "/Users/annasintsova/git_repos/spatial_dynamics_of_gene_expression_in_response_to_T6SS_attack/data/counts/reverse/"
gff = "/Users/annasintsova/git_repos/spatial_dynamics_of_gene_expression_in_response_to_T6SS_attack/data/ref/HI4320_PATRIC.gff"
```

```python
tpms = counts.normalize_counts_to_tpm(cf, gff)
df = pd.DataFrame.from_dict(tpms)
tags = []
for f in df.index:
    try:
        tag = pids[f]
    except KeyError:
        tag = f
    tags.append(tag)
df["tags"] = tags
df = df.set_index("tags")

fn = os.path.join(os.path.dirname(cf), "TPMs.csv")
df.to_csv(fn)

df.head()
```
